```python
marks = (90, 82, 66, 76, 33)

maximum = max(marks)
print("Maximum marks:", maximum)

minimum = min(marks)
print("Minimum marks:", minimum)

total_count = len(marks)
print("Total number of marks:", total_count)

total_sum = sum(marks)
print("Sum of marks:", total_sum)

average = sum(marks) / len(marks)
print("Average:", average)

sorted(marks)
print("Marks in ascending order:", marks)
```

```python
# Student exam result
result = ("Mathematics", 82, "A")

# Unpack the tuple
subject, marks, grade = result

# Display the information
print("Subject:", subject)
print("Marks Obtained:", marks)
print("Grade:", grade)

# Check for distinction
if marks >= 75:
    print("Distinction")
```

```python
attendance = ("P", "P", "A", "P", "P", "P", "A")

print("=== Employee Attendance Record ===\n")

# Display the attendance record
print("Weekly Attendance:", attendance)
print()

# Count total number of present days
present_days = attendance.count("P")
print("Total Present Days:", present_days)

# Count total number of absent days
absent_days = attendance.count("A")
print("Total Absent Days:", absent_days)

print()

# Check if employee was absent at least once and display message
if "A" in attendance:
    print(f"Message: Employee was present for {present_days} days and absent for {absent_days} days during the week.")
else:
    print(f"Message: Employee was present for all {present_days} days. Perfect attendance!")


```
